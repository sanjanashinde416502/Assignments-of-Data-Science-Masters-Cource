```python
# Q.2 What is DDL ? Explain why CREATE, DROP, ALTER , and TRUNCATE are used with an example.
# Ans:
# What is DDL :
#     DDL stands for Data Definition Language, which is a subset of SQL (Structured Query Language) used to define and modify the structure 
# of a database. DDL statements are used to create, modify, and delete database objects such as tables, indexes, and constraints.

# The four commonly used DDL statements are:

# 1.CREATE: 
#    This statement is used to create a new database object such as a table, index, or view. For example, the following SQL code creates a 
# new table named "employees" with columns for "employee_id," "employee_name," and "employee_salary":
# SQL code:
# CREATE TABLE employees (
#     employee_id INT PRIMARY KEY,
#     employee_name VARCHAR(50),
#   employee_salary DECIMAL(10,2)
# );

# 2.DROP: 
#    This statement is used to remove an existing database object such as a table or view. 
# For example, the following SQL code drops the "employees" table:
# sql code:
# DROP TABLE employees;

# 3.ALTER:
#   This statement is used to modify the structure of an existing database object such as a table or column.
# For example, the following SQL code adds a new column named "employee_department" to the "employees" table:
# sql code:
# ALTER TABLE employees ADD employee_department VARCHAR(50);

# 4.TRUNCATE: 
#   This statement is used to delete all the data from a table, while keeping the table structure intact.
# For example, the following SQL code removes all data from the "employees" table:
# sql code:
# TRUNCATE TABLE employees;
```

```python
# Q.3 What is DMl ? Explain  INSERT ,UPDATE,and DELETE with an example.
# Ans:
# What is DML:
#    DML stands for Data Manipulation Language, which is a subset of SQL (Structured Query Language) used to manipulate data in a relational database 
# management system (RDBMS). DML commands are used to add, modify, or delete data in the database.

# Following  are  DML commands:

# 1.INSERT: 
#    The INSERT command is used to add new data into a table. 
# The basic syntax for the INSERT command is:
# INSERT INTO table_name (column1, column2, column3, ...)VALUES (value1, value2, value3, ...);

# For example:
#   let's say we have a table named "students" with the columns "id", "name", and "age". We can insert a new record into the table using the following
# command:

# INSERT INTO students (id, name, age)
# VALUES (1, 'John', 20);

# This command will add a new record with id = 1, name = 'John', and age = 20 to the students table.

# 2.UPDATE:
#    The UPDATE command is used to modify existing data in a table. 
# The basic syntax for the UPDATE command is:
# UPDATE table_name
# SET column1 = value1, column2 = value2, ...
# WHERE condition;

# For example:
#    let's say we want to update the age of the student with id = 1 from 20 to 21. We can use the following command:

# UPDATE students
# SET age = 21
# WHERE id = 1;

# This command will modify the age of the student with id = 1 to 21.

# 3.DELETE:
#   The DELETE command is used to delete existing data from a table. 
# The basic syntax for the DELETE command is:
# DELETE FROM table_name
# WHERE condition;

# For example:
#    let's say we want to delete the record of the student with id = 1 from the students table. We can use the following command:

# DELETE FROM students
# WHERE id = 1;

# This command will delete the record of the student with id = 1 from the students table.
```

```python
# Q.5 Explain Primary Key and Foreign Key.
# Ans:
#    In database design, a primary key and foreign key are both important concepts that help establish relationships between tables.

# Primary Key :
#    A primary key is a column or set of columns in a table that uniquely identifies each record in the table. It must have a unique value for each 
# record, and cannot contain null values. Primary keys are used as a reference by other tables, which create relationships with the primary key of 
# another table. 
# For example:
# in a table of customers, the primary key might be the "customer_id" column, which uniquely identifies each customer in the table.

# Foreign Key :
#    A foreign key is a column or set of columns in a table that refers to the primary key of another table. It establishes a relationship between two
# tables and is used to enforce referential integrity, which ensures that data is consistent and accurate across tables. For example, if you have a 
# table of orders and a table of customers, you might include a foreign key in the orders table that references the primary key in the customers table.
# This would ensure that each order is associated with a valid customer in the customer table.
```

```python
# Q.7 Give the order of execution of SQL clauses in an SQL query.
# Ans:
# In an SQL query, the clauses are executed in the following order:

# 1.FROM: This clause specifies the table or tables that the query will retrieve data from.

# 2.WHERE: This clause is used to filter the data returned by the query based on a specified condition. It is executed after the FROM clause.

# 3.GROUP BY: This clause is used to group the data returned by the query into sets based on specified columns. It is executed after the WHERE clause.

# 4.HAVING: This clause is used to filter the data returned by the query based on a specified condition for a group created by the GROUP BY clause. 
#           It is executed after the GROUP BY clause.

# 5.SELECT: This clause specifies the columns to be included in the result set. It is executed after the HAVING clause.

# 6.DISTINCT: This clause is used to remove duplicates from the result set. It is executed after the SELECT clause.

# 7.ORDER BY: This clause is used to sort the result set based on specified columns. It is executed after the SELECT and DISTINCT clauses.

# 8.LIMIT: This clause is used to limit the number of rows returned by the query. It is executed after the ORDER BY clause.
```
