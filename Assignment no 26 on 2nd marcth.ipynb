```python
# Q.2 What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
#     plot a scatter plot.
# Ans:
#    What is a scatter plot? 
#        A scatter plot is a type of data visualization that shows the relationship between different variable.
#    It uses points(or dots) to represent individual data elements , with one variable shown on the x-axis and another on the y-axis.

# Note: Also add title, xlabel, and ylabel to the plot.
# ploting scatter plot:
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatter Plot')
```

```python
# Q.3 Why is the subplot() function used? Draw four line plots using the subplot() function.
# Ans:
#   Why is the subplot() function used?
#       The 'subplot()'function is used to create multiple plots (subplots) within a single figure in Python. This allows you to display different visualizations or data comparisons side
#   by side, making it easier to analyze and compare multiple datasets or aspects of the same data ina concise and organized manner.

# Use the following data:
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Data for line plot
x=np.array([0,1,2,3,4,5])

# data for each line
y1=np.array([0,100,200,300,400,500])
y2=np.array([50,20,40,20,60,70])
y3=np.array([10,20,30,40,50,60])
y4=np.array([200,350,250,550,450,150])

# Create a 2x2 subplot layout
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("Line 1 plot")

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('Line 2 plot')

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('Line 3 plot')

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('Line 4 plot')

plt.tight_layout()
plt.show()
```

```python
# Q.4 What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
# Ans:
#    What is a bar plot:
#         A bar plot is also known as bar chart , is a graphical representation of data using rectangular bars of equal width.It is commonly used to display and compare categorical data.
#    The length of each bar is proportional to the value it represents, making it easy to visually compare the data points.

#    Why is it used?
#         1.For comparing different categories or groups of data
#         2.For displaying discrete data, where each category represents a distinct entity or lable
#         3.For showing the distribution of a categorical variable 
#         4.For highlighting treads or patterns in the data

# creating vertical and horizontal bar plots
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# vertical bar plots
plt.figure(figsize=(5,3))
plt.bar(company,profit)
plt.title('Vertical bar plots')
plt.xlabel('Company')
plt.ylabel('Profit')

# horizontal bar plot
plt.figure(figsize=(5,3))
plt.barh(company,profit)
plt.title('Horizontal bar plots')
plt.xlabel('Company')
plt.ylabel('Profit')
```

```python
# Q.5 What is a box plot? Why is it used? Using the following data plot a box plot.
# Ans:
#    What is a box plot:
#        A box plot is a graphical representation of the distribution of numerical data through 5 key summary statistics. The minimum value,first quartile(Q1),median (Q2),third 
#     quartile(Q3), and maximum value.
#    The box in the plot represents the interquartile range(IQR) which is ranges between first quartile and third quartiles (Q1-Q3)

#    Why is it used?
#        1.for indentifying the central tendancy (median) and spread(IQR) of the data
#        2.for detecting skewness or symmetry in the data distribution
#        3.for comparing the distributions of multiple datasets side-by-side
#        4.for identifying potential outliers in the data.

# ploting bar plot
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data=[box1,box2]
lables=['Box1','Box2']

plt.figure(figsize=(5,4))
plt.boxplot(data,labels=lables, vert=False,patch_artist=True)
plt.xlabel('Values')
plt.ylabel('Boxs')
plt.grid(axis='x',linestyle='--',alpha=0.7)
plt.show()
```
