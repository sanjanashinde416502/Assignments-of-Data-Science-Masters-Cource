```python
# 2.OverflowError: 
#        This error is raised when an arithmetic operation exceeds the maximum representable value for a numeric type. 
# For example:

j=5.0
try:
    for i in range(1,1000):
        j**i
        
except ArithmeticError as e:
    print(e)
```

```python
# Explaination:
#      In this case, adding 1 to the maximum representable value for a numeric type (sys.maxsize) causes an overflow, which Python detects
# and raises an OverflowError exception.
```
