```python
# Q.1 Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
# Ans:
# following  are the plots that we can plot using seaborn:
#  1.Scatter Plot: 
#         Scatter plot used to visulize the relationship between two numerical variables.They help us to usnderstand patterns,correlations and potential outliers in the data

#  2.Bar Plot:
#         Bar plots are used to visulize and compare the distribution of categorical data.This is useful for showing group comparisons.

#  3.Histogram:
#         A histogram is a graph that shows frequency distribution of set of continues data.

#  4.Line Plot:
#         A line plot is the simplest plot in all ploting types . This is used to visulize the trend of a numerical variable over a continues interval.

#  5.Heatmap:
#         A heatmap are great for visulizing relationships between two categorical variables using colors.A heatmap is a graphical representation of data using colors to indicate the values of the data in a 2D matrix 
```

```python
# Q.2 Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.
#     Note: timepoint, signal, event, and region are columns in the fmri dataset.
# Ans:
import seaborn as sns
df=sns.load_dataset("fmri")
sns.lineplot(data=df,x=df['timepoint'],y=df['signal'],hue="event",style="region")
sns.despine()
```

```python
# Q.3 Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.
#    Note: pclass, age, and fare are columns in the titanic dataset.
# Ans:
df = sns.load_dataset("titanic")

# Create box plots using Seaborn
sns.set(style="whitegrid")  # Set the plotting style

# Box plot for 'age' based on 'pclass'
sns.boxplot(data=df, x=df['pclass'], y=df['age'])
sns.despine()  # Remove spines from the plot

# Box plot for 'fare' based on 'pclass'
sns.boxplot(data=df, x=df['pclass'], y=df['fare'])
sns.despine()  # Remove spines from the plot
```

```python
# Q.4 Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hueparameter for the 'cut' column of the diamonds dataset.
# Ans:
df=sns.load_dataset('diamonds')
sns.set(style='dark')
sns.histplot(data=df,x=df['price'],hue='cut',multiple='stack')
sns.despine()
```

```python
# Q.5 Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" columnof the iris dataset.
# Ans:
df=sns.load_dataset('iris')
sns.pairplot(data=df,hue="species")
```

```python
# Q.6 Use the "flights" dataset from seaborn to plot a heatmap.
# Ans:
# Load the "flights" dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to create a matrix suitable for a heatmap
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")

# Create a heatmap using Seaborn
sns.set(style="whitegrid")  # Set the plotting style
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
```
