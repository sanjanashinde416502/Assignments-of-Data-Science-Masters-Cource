```python
# Q.2 State and explain the features of MonogoDB.
# Ans:
# MonogoDB:
#      MongoDB is a popular NoSQL database management system that is used for its scalability, flexibility, and high-performance capabilities.
# Some of the key features of MongoDB are:

# 1.Document-Oriented: 
#      MongoDB is a document-oriented database, which means that data is stored in a flexible, semi-structured document format instead of tables with
# rows and columns. This allows for more efficient and dynamic handling of complex data structures.

# 2.Dynamic Schema: 
#      MongoDB allows for dynamic schema design, which means that documents within a collection can have different structures and fields. This makes 
# it easier to adapt to changing data requirements and to scale the database as needed.

# 3.Scalability: 
#      MongoDB is designed for high scalability and can handle large volumes of data and high traffic. It provides features such as horizontal scaling,
# sharding, and load balancing to ensure that the database can handle increased data volumes and user requests.

# 4.High Availability:
#      MongoDB provides high availability through its replica set feature, which automatically creates multiple copies of data across different 
# servers. This ensures that data is always available even in the event of a server failure.

# 5.Querying and Indexing: 
#      MongoDB provides a powerful query language and indexing capabilities to make it easy to retrieve and search for data. Its indexing capabilities
# allow for faster search and retrieval of data, making it ideal for applications that require real-time processing.

# 6.Security:
#      MongoDB provides security features such as authentication, access control, encryption, and auditing to ensure that data is secure and protected.
```
