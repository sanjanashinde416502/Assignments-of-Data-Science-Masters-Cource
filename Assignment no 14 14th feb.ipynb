```python
# Q.1 What is multithreading in python ? Why is it used ? Name the module used to handle thresds in python .
# Ans:
# What is multithreading:
#       Multithreading is the ability of a program to execute multiple threads (or parallel streams of execution) concurrently. In Python, 
# multithreading is a way to achieve concurrency and parallelism by running multiple threads at the same time, each performing a specific
# task.

# Why is is used :
#       Multithreading is used to improve the performance of a program by making better use of available resources, such as CPU cores. 
# By executing multiple threads in parallel, a program can perform multiple tasks simultaneously, which can significantly improve its 
# performance.

# Name of module used :
#   The threading module is used to handle threads in Python. This module provides a way to create, start, and control threads.
```

```python
# Q.2 Why threading module used ? Write the use of the following functions.
#     1.activeCount()
#     2.currentThread()
#     3.enumerate()

# Ans:
# Why threading module is used :
#     The threading module in Python is used for creating and managing threads in a Python program. Threads are a way of achieving 
# parallelism in programs by dividing the program into smaller threads of execution that can run simultaneously on different processors or
# cores.

# Use of the following functions:

# 1.activeCount():
#     This function returns the number of active threads in the current process. It can be useful for debugging and monitoring purposes, 
# as it allows you to keep track of how many threads are running at any given time.

# 2.currentThread():
#     This function returns a reference to the current thread object. The thread object represents the thread that is currently executing 
# the code. It can be useful for obtaining information about the current thread, such as its name or identifier.

# 3.enumerate(): 
#     This function returns a list of all thread objects that are currently active. It can be useful for iterating over all threads in a 
# program and performing operations on them, such as joining or terminating them. The returned list includes the main thread as well as
# any other active threads that have been started with the threading module.
```

```python
# Q.4 Write a python program to create two threads. Thread one must print the list of squares and thread two must print list of cubes.
# Ans:
import threading
def square():
    for i in range(1,11):
        print("Square of %d is %d" %(i,i**2))
        
def cube():
    for i in range(1,11):
        print("Cube of %d is %d" %(i,i*i*i))
        
t1=threading.Thread(target=square)
t2=threading.Thread(target=cube)


print("Square of first 10 numbers are:\n")
t1.start()
print("\n")
print("Cube of first 10 numbers are:\n")
t2.start()

t1.join()
t2.join()
    
```

```python
# Q.5 State advantage and disadvantage of multithreading.
# Ans:
# Advantages:

# 1.Increased Performance:
#       Multithreading can improve the performance of an application by allowing multiple threads to execute simultaneously on a multi-core processor. 
# This can result in faster processing of tasks and overall better application performance.

# 2.Resource Sharing:
#      Threads within the same process share the same memory and resources, which can lead to more efficient use of system resources and less overhead
# compared to running multiple processes.

# 3.Responsiveness: 
#      Multithreading can make an application more responsive to user input by allowing user interface operations to execute on a separate thread,
# which prevents the interface from becoming unresponsive during long-running operations.

# 4.Scalability:
#      Multithreading can improve the scalability of an application by allowing it to handle more users or requests without having to increase the 
# number of processes.

# Disadvantages:

# 1.Complexity:
#      Multithreading adds complexity to the application design and programming, as threads must be carefully synchronized to avoid race conditions
# and other issues.

# 2.Synchronization Overhead:
#      Synchronization between threads can add overhead and reduce performance, as threads must wait for access to shared resources.

# 3.Difficult Debugging: 
#      Debugging multithreaded applications can be challenging, as bugs may occur only under specific timing conditions, and race conditions can be 
# hard to reproduce.

# 4.Resource Contentions:
#      When multiple threads access the same resources, such as files or network connections, they can contend for those resources, which can result 
# in reduced performance or deadlocks.

```
