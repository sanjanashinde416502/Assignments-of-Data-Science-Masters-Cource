```python
# Q.2 What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
# Ans:
# What are glyphs in Bokeh:
#    In Bokeh , Glyphs are visual markers that can be added to plots to represent data points.
# Glyphs are the building blocks for creating visulizations in Bokeh and they include various shapes such as circles , squares , lines and more

from bokeh.plotting import figure, show
from bokeh.io import output_file

# Create some example data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 8]

# Create a Bokeh figure
p = figure(title='Scatter Plot Example', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add circles as glyphs to the figure
p.circle(x, y, size=5, color='navy', alpha=0.5)

# Specify the output file
output_file('scatter_plot.html')

# Show the plot
show(p)
```

```python
# Q.3 How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
# Ans: 
# Customizing the appearance of a Bokeh plot involves modifying various elements such as axes, title, legend, colors, and more. Bokeh provides a flexible and extensive set of options for customization. Below are some common customization tasks for different elements of a Bokeh plot:

# 1. Plot Title:
#     We can set the title of the plot using the 'title' attribute
from bokeh.plotting import figure, show

plot = figure(title="My Custom Title")

# 2. By Axes:
#     Customize the x and y-axis labels,tricks and their properties
plot.xaxis.axis_label = "X-axis Label"
plot.yaxis.axis_label = "Y-axis Label"

plot.xaxis.major_label_text_color = "red"
plot.yaxis.major_label_text_font_size = "14pt"

# 3 Legend:
#     If our plot has mutiple glyphs or lines , we can add a legend.
plot.line(x, y, legend_label="Line 1")
plot.circle(x, y, legend_label="Circle")

plot.legend.title = "Legend Title"
plot.legend.label_text_font_size = "12pt"
```
