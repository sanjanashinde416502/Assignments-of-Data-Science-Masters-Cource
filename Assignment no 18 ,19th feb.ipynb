```python
# Q.3 What is APP routing in Flask ? Why do we use app routes ?
# Ans:
# What is APP routing:
#     In Flask, an "app route" refers to a URL endpoint that responds to HTTP requests made to a Flask application. Flask uses a decorator
# called @app.route() to define app routes in your Flask application.

# Why we use app routes :
#     App routes are used in Flask to define the behavior of your application in response to incoming requests. By defining app routes, you
# can control what happens when a user clicks a link or submits a form on your website. Flask provides a simple and flexible way to define 
# app routes that makes it easy to build complex web applications.
```
