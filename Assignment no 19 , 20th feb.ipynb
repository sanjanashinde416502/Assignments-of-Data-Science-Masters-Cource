```python
# Q.1 Explain GET and POST methods.
# Ans:
# GET and POST methods.
#       The GET and POST methods are two of the most common HTTP methods used for communicating between clients (such as web browsers) and 
# servers.

# GET method :
# 1.The GET method is used to retrieve data from a server. 
# 2.When a client sends a GET request, it typically includes a URL (Uniform Resource Locator) that specifies the location of the desired 
#   resource. 
# 3.The server then responds with the requested data, which may include HTML pages, images, videos, or other types of content.
# 4.The GET method is considered "safe" because it does not modify any data on the server and can be repeated multiple times without causing
#   any side effects.

# POST method :
# 1.The POST method, on the other hand, is used to send data to a server to create or update a resource. 
# 2.When a client sends a POST request, it typically includes a message body that contains the data to be sent to the server. 
# 3.The server then processes the data and sends a response back to the client. The POST method is considered "unsafe" because it can modify
#   data on the server and may have side effects, such as creating or deleting resources.

#      In summary, the GET method is used for retrieving data from a server, while the POST method is used for sending data to a server to
# create or update a resource.
```

```python
# Q.2 Why is request used in Flask?
# Ans:
#     In Flask, the request object is used to access incoming request data such as form data, query parameters, and the contents of POST 
# and PUT requests. It is a global variable that is available within the context of a request and holds all of the information related to
# the client's request.

#    Using the request object, you can access the data that the client has sent to the server as part of their request. For example, you
# can retrieve form data using the request.form attribute, query parameters using the request.args attribute, and the contents of POST and 
# PUT requests using the request.data attribute.

#    You can also use the request object to access other information related to the client's request, such as the HTTP method used to make 
# the request (request.method), the URL of the request (request.url), and the headers of the request (request.headers).
```

```python
# Q.3. Why is redirect() used in Flask?
# Ans:
#      redirect() is a function provided by the Flask framework that allows you to redirect a user to a different URL from within a Flask 
# view function. It's commonly used to redirect a user to a different page after they have submitted a form or performed some other action
# on a website.

#      The redirect() function works by sending an HTTP response with a "302 Found" status code, which tells the user's web browser to send 
# a new request to the URL that you specify as an argument to the redirect() function.
```
